```python
sns.boxplot(data =df, x= "Sex", y = "Age", hue= df["Survived"])
```

```python
df[df["Survived"]==0]["Age"]  #those who did not survive.
```

```python
df[df["Survived"]==0]["Age"].max()
```

```python
df[df["Survived"]==0]["Age"].min()
```

```python
df[df["Survived"]==1]["Age"]   #those who survive.
```

```python
df[df["Survived"]==1]["Age"].max()
```

```python
df[df["Survived"]==1]["Age"].min()
```

```python
sns.distplot(df[df["Survived"]==0]["Age"], hist = False)
sns.distplot(df[df["Survived"]==1]["Age"], hist = False)
```

```python
survived_0_age = df[df["Survived"] == 0]["Age"]

sns.kdeplot(data=survived_0_age, fill=False)

survived_1_age = df[df["Survived"] == 1]["Age"]

sns.kdeplot(data=survived_1_age, fill=False)
```
